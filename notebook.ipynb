```python
from openai import OpenAI

client = OpenAI()
```

```python
messages = [
    # {"role": "system", "content": "You are an expert Python programmer, designing retro and small games. You also have graphical abilities, and draw SVG images of all assets required by your games."},
    {"role": "user", "content": "Write the classical Chrome Dino game using PyGame. Make sure to keep some randomness to make it different every time it launches, and draw all SVG images required too. Write all Python code in a single file"},
]

response = client.chat.completions.create(
    model='o1-preview',
    messages=messages,
    # temperature=0.3
)

response.choices[0].message.content
```
